```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv('../data/users.csv')
df.head()
```

```python
data = df[df['campus'] == 1]
data.head()
```

```python
data.drop(['campus', 'id'], axis=1, inplace=True)
data.head()
```

```python
data.describe()
```

```python
max_values = {
    '00': 125,
    '01': 125,
    '02': 105,
    '03': 100,
    '04': 100,
    '05': 115,
    '06': 125,
    '07': 100,
    '08': 100,
    '09': 100,
    'r00': 119,
    'r01': 125,
    'final': 125,
}
mask = data['r00'] > max_values['r00']
for k, v in max_values.items():
    mask = mask | (data[k] > v)

mask.value_counts()
```

```python
new_data = data.drop(data[mask].index)
data.shape, new_data.shape, 'Dropped: ' + str(data.shape[0] - new_data.shape[0])
```

```python
y = new_data['final']
X = new_data.drop('final', axis=1)
X.shape, y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python
with open('../pickle/paris_campus.pickle', 'wb+') as f:
    pickle.dump(X_train, f)
    pickle.dump(X_test, f)
    pickle.dump(y_train, f)
    pickle.dump(y_test, f)
    f.close()
```
